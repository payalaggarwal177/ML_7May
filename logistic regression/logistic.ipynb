## Step 3 : Seprate X and Y species
species ~ sep_len, sep_wid, pet_len, pet_wid

```python
X = df.drop(columns = 'species')
Y = df[['species']]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

## Step 4: Apply Preprocessing on X

```python
X.dtypes
```

Because al features are numerical i will create a numerical pipeline only

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps = [('impute', SimpleImputer(strategy='mean')),
                             ('scaler', StandardScaler())]).set_output(transform = 'pandas')
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

## Step 5 : Apply train test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size = 0.33, 
                                                random_state = 21)
```

```python
xtest.head()
```

```python
ytrain.head()
```

```python
ytest.head()
```

```python
xtest.shape
```

## Step 6: Bluiding the logistic Regrssion MOdel


```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

```python
model.classes_
```

```python
model.intercept_
```

```python
model.coef_
```

```python
c = X.corr()
c
```

```python
import seaborn as sns
sns.heatmap(data=c, annot =True,fmt=' .2f' )
```

## Step 7: Model evalution

```python
# train accuracy
model.score(xtrain, ytrain)
```

```python
# test accuracy
model.score(xtest, ytest)
```

```python
# Confusion matrix
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(model, xtrain, ytrain)
```

```python
ConfusionMatrixDisplay.from_estimator(model, xtest, ytest)
```

## Cross validate result on train data

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, xtrain, ytrain, cv=5, scoring='f1_macro')
scores
```
